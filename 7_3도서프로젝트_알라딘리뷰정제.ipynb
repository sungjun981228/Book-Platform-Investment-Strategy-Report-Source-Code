```python
import pandas as pd
```

## 100자평과 그냥 리뷰 합치기

```python
# 불러오기

al_bdf1 = pd.read_csv("[알라딘]100자리뷰_베스트셀러.csv", encoding='utf-8-sig')
al_bdf2 = pd.read_csv("[알라딘]리뷰_베스트셀러.csv", encoding='utf-8-sig')
```

```python
# 병합
al_review_best = pd.concat([al_bdf1, al_bdf2], ignore_index=True)

# isbn 기준으로 정렬
al_review_best = al_review_best.sort_values(by='isbn').reset_index(drop=True)
```

```python
al_review_best
```

## 열이름 변경

```python
al_review_best.rename(columns={'star_rating': 'rating'}, inplace=True)
al_review_best.rename(columns={'review_content': 'review'}, inplace=True)
```

```python
al_review_best
```

## 닉네임 지우기

```python
al_review_best = al_review_best.drop(['nickname'], axis = 1)
```

## 열 순서 변경

```python
al_review_best = al_review_best[['isbn', 'review', 'rating']]
```

```python
al_review_best
```

## 중복 행 지우기

```python
al_review_best = al_review_best.drop_duplicates()
```

## 누락 행, 없음 행 지우기

```python
al_review_best = al_review_best[al_review_best['review'].fillna("").str.strip() != "없음"]
```

```python
al_review_best
```

## 숫자로만 이루어진 리뷰 삭제

```python
al_review_best = al_review_best[~al_review_best['review'].fillna("").astype(str).str.match(r'^\d+$')]
```

```python
al_review_best
```

## 리뷰가 아닌 내용들, 이모지, 특문, 공백 삭제

```python
import re

def 리뷰정리(text):
    if pd.isnull(text):
        return text
    text = str(text)
    
    # 날짜+공감+댓글 패턴 제거
    text = re.sub(r'\d{4}-\d{2}-\d{2}\s공감\s\(\d+\)\s댓글\s\(\d+\)', '', text)

    # 줄바꿈 문자 제거
    text = text.replace('\n', ' ').replace('\r', ' ')

    # 이모지 제거 (유니코드 기반)
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text)

    # 특수문자 제거 (한글, 영문, 숫자, 공백만 남김)
    text = re.sub(r'[^가-힣a-zA-Z0-9\s]', '', text)

    # 공백 여러 개 → 하나로 압축
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
al_review_best['review'] = al_review_best['review'].apply(리뷰정리)
```

## Thanks to 공감 지우기

```python
al_review_best[al_review_best['review'].fillna("").str.contains(r'Thanks to 공감')]
```

```python
def remove_공감(text):
    if pd.isnull(text):
        return text
    return str(text).replace("Thanks to 공감", "").strip()
```

```python
al_review_best['review'] = al_review_best['review'].apply(remove_공감)
```

```python
al_review_best[al_review_best['rating'] == '없음']
```

```python
# 문자열 '없음'을 NaN으로 처리하고 숫자로 변환
al_review_best['rating'] = pd.to_numeric(al_review_best['rating'], errors='coerce')

# isbn별 평균 구하기
isbn_rating_mean = al_review_best.groupby('isbn')['rating'].mean()

# isbn별 평균값을 결측치에 채우기
al_review_best['rating'] = al_review_best.apply(
    lambda row: isbn_rating_mean[row['isbn']] if pd.isnull(row['rating']) else row['rating'],
    axis=1
)
# 여기서도 NaN이면 isbn 전체가 NaN인거임
```

```python
al_review_best = al_review_best[~al_review_best['rating'].isna()]
```

```python
al_review_best['rating'] = al_review_best['rating'].astype(int)
```

```python
al_review_best
```

# 비베스트셀러
