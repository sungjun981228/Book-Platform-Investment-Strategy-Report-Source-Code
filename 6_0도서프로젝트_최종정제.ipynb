```python
import re
normalize_rules = {
    r'miraen|아이세움': '미래엔',
    r'오렌지미디어': '오렌지미디어',
    r'뜨인돌': '뜨인돌',
    r'인티.?앤': '인티앤',
    r'rhk|알에이치코리아': '알에이치코리아',
    r'한경bp|한국경제신문': '한국경제신문',
    r'휴먼큐브|아이휴먼': '휴먼큐브',
    r'카시오페아': '카시오페아',
    r'포레스트북스|콘텐츠그룹 포레스트|리프': '포레스트북스',
    r'청어람미디어|나무의말': '청어람미디어',
    r'필름|feelm|cloudia|클라우디아': '필름',
    r'블루홀식스|블루홀6': '블루홀식스',
    r'길벗|더 퀘스트': '길벗',
    r'웅진씽크빅|리더스북|갤리온|웅진지식하우스|웅진주니어': '웅진씽크빅',
    r'21세기북스|북이십일': '북이십일',
    r'웨일북|whale books': '웨일북',
    r'사계절': '사계절'
}

# 정규화 함수 정의
def normalize_publisher(publisher):
    if pd.isna(publisher):
        return publisher
    publisher_lower = publisher.lower()  # 대소문자 무시
    for pattern, replacement in normalize_rules.items():
        if re.search(pattern, publisher_lower):
            return replacement
    return publisher  # 규칙에 없으면 원본 유지

# 정규화된 컬럼 추가
df['PUBLER'] = df['PUBLER'].apply(normalize_publisher)
```

```python
import re
import pandas as pd

normalized = []

for name in df['AUTHOR']:
    # 문자열이 아닌 경우 (예: NaN, float 등) 처리
    if not isinstance(name, str):
        normalized.append('')
        continue

    # 1. 대괄호 앞 제거
    name = name.split('[')[0].strip()

    # 2. "지은이: " 형태면 콜론 뒤 이름만 남기기
    if '지은이:' in name:
        name = name.split('지은이:')[-1].strip()
    
    # 3. "풀어씀", "씀", "쓰고" 등이 포함되면 그 앞까지만 남기기
    for marker in ['풀어씀', '씀', '쓰고']:
        if marker in name:
            name = name.split(marker)[0].strip()
    
    # 4. 역할 키워드 나오기 전까지 자르기
    pattern = r"(글|그림|지음|원저|저|엮음|노랫말)"
    name = re.split(pattern, name)[0].strip()
    
    normalized.append(name)

df['AUTHOR'] = normalized

df.loc[df['AUTHOR'].str.contains('원작', na=False), 'AUTHOR'] = df['AUTHOR'].str.extract(r'^(.*?)\s*원작')
```

```python
import re

normalized = []

for name in df['AUTHOR']:
    # 0. 비문자열 처리
    if not isinstance(name, str):
        normalized.append('')
        continue

    # 1. by로 시작하면 by 뒤만 남김
    name = name.strip()
    if name.lower().startswith("by "):
        name = name[3:].strip()

    # 2. 역할 키워드 앞까지 자르기
    name = re.split(r"(시|역주|편|장편무협소설)", name)[0].strip()

    # 3. 외/外 정보 제거
    name = re.split(r"(외\s*\d*인|外)", name)[0].strip()

    # 4. 숫자 + '명' 제거
    name = re.split(r"\d+명", name)[0].strip()

    normalized.append(name)

df['AUTHOR'] = normalized

```

```python
import re

def clean_by_author(author):
    if not isinstance(author, str):
        return author  # 비문자열은 그대로 반환

    # 'by' 또는 '[by]'로 시작하거나 포함되어 있을 경우
    match = re.search(r'\[?by\]?\s*(.*)', author, re.IGNORECASE)
    if match:
        return match.group(1).strip()  # 'by' 뒤의 내용만 반환
    return author.strip()  # 없으면 원래 값

# 적용
df['AUTHOR'] = df['AUTHOR'].apply(clean_by_author)
```

```python
import re

def clean_author(author):
    if not isinstance(author, str):
        return author  # 비문자열은 그대로

    name = author.strip()

    # 1. [공자 지음] → 괄호 제거
    name = re.sub(r'[\[\]]', '', name)

    # 2. "by" 또는 "[by]" → by 뒤 이름만
    match_by = re.search(r'\[?by\]?\s*(.*)', name, re.IGNORECASE)
    if match_by:
        return match_by.group(1).strip()

    # 3. "역할: 이름" → ":" 뒤 이름만
    if ':' in name:
        name = name.split(':', 1)[-1].strip()

    # 4. "외", "외 2인", "外", "9명" 등 제거
    name = re.split(r'(외\s*\d*인|외|外|\d+명)', name)[0].strip()

    # 5. 역할 키워드 제거 (지음, 엮은이, 글, 저, 지은이 등)
    role_pattern = r"(지음|엮은이|글|저|지은이)"
    name = re.split(role_pattern, name)[0].strip()

    return name

# 적용
df['AUTHOR'] = df['AUTHOR'].apply(clean_author)
```

```python
import re
normalize_rules = {
    r'miraen|아이세움': '미래엔',
    r'오렌지미디어': '오렌지미디어',
    r'뜨인돌': '뜨인돌',
    r'인티.?앤': '인티앤',
    r'rhk|알에이치코리아': '알에이치코리아',
    r'한경bp|한국경제신문': '한국경제신문',
    r'휴먼큐브|아이휴먼': '휴먼큐브',
    r'카시오페아': '카시오페아',
    r'포레스트북스|콘텐츠그룹 포레스트|리프': '포레스트북스',
    r'청어람미디어|나무의말': '청어람미디어',
    r'필름|feelm|cloudia|클라우디아': '필름',
    r'블루홀식스|블루홀6': '블루홀식스',
    r'길벗|더 퀘스트': '길벗',
    r'웅진씽크빅|리더스북|갤리온|웅진지식하우스|웅진주니어': '웅진씽크빅',
    r'21세기북스|북이십일': '북이십일',
    r'웨일북|whale books': '웨일북',
    r'사계절': '사계절'
}

# 정규화 함수 정의
def normalize_publisher(publisher):
    if pd.isna(publisher):
        return publisher
    publisher_lower = publisher.lower()  # 대소문자 무시
    for pattern, replacement in normalize_rules.items():
        if re.search(pattern, publisher_lower):
            return replacement
    return publisher  # 규칙에 없으면 원본 유지

# 정규화된 컬럼 추가
df['PUBLER'] = df['PUBLER'].apply(normalize_publisher)

df['PUBLER'] = df['PUBLER'].str.replace(r'[:\(\[\{].*$', '', regex=True).str.strip()
```
