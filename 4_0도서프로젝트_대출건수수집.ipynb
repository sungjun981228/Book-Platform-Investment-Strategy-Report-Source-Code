# 2020 ~ 2024 ISBN별 대출건수 통합

```python
import pandas as pd

df2020 = pd.read_csv('대출건수_2020.csv')
df2021 = pd.read_csv('대출건수_2021.csv')
df2022 = pd.read_csv('대출건수_2022.csv')
df2023 = pd.read_csv('대출건수_2023.csv')
df2024 = pd.read_csv('대출건수_2024.csv')

df_concat = pd.concat([df2020, df2021, df2022, df2023, df2024], ignore_index=True)
```

```python
df_concat['isbn13'].duplicated().sum()
```

```python
df_concat.to_csv('대출건수_통합.csv', index=False)
```

```python
# ISBN별로 loan_count 합산
grouped_df = df_concat.groupby('isbn13', as_index=False)['loan_count'].sum()

# 결과 확인
print(grouped_df.head())
```

```python
grouped_df.rename(columns={'isbn13': 'ISBN'}, inplace=True)
```

```python
df = pd.read_csv('베스트셀러_통합.csv')
df
```

```python
# grouped_df: ISBN별 loan_count가 합쳐진 DataFrame이라고 가정
isbn_to_count = dict(zip(grouped_df['ISBN'], grouped_df['loan_count']))
df['LOAN_COUNT'] = df['ISBN'].map(isbn_to_count)
```

```python
df[df['LOAN_COUNT'].isna()]
```

```python
df.to_csv('베스트셀러_통합1차.csv', index=False)
```

# API 호출제한으로 대출건수 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd

# ChromeOptions 설정
options = webdriver.ChromeOptions()
# options.add_argument('--headless')  # 크롬창 숨기기 (필요시 주석 해제)
options.add_argument('--start-maximized')  # 크롬창 최대화
options.add_experimental_option('detach', True)  # 코드 종료 후 창 유지
options.add_argument("--disable-blink-features=AutomationControlled")  # 탐지 우회

# 드라이버 실행
driver = webdriver.Chrome(options=options)

driver.get('https://www.data4library.kr/bookV?seq=1241509')

for j in df['ISBN']:
    try:
        # 검색창에 ISBN 입력
        search_box = driver.find_element(By.XPATH, '//*[@id="headerSrchText"]')
        search_box.clear()
        search_box.send_keys(j)
        time.sleep(1)
        search_box.send_keys(Keys.ENTER)
        time.sleep(2)  # 페이지 로딩 대기

        # 검색 결과 확인
        try:
            driver.find_element(By.XPATH, '//*[@id="sb-site"]/section/div[2]/div/div/div[3]/div[2]/table/tbody/tr/td[2]/a')
        except:
            print(f"검색 결과 없음 (ISBN: {j})")
            continue

        # 페이지 파싱
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        word = soup.find('td', 'br_none').text
        df.loc[df['ISBN'] == j, 'LOAN_COUNT'] = word
        print(f"{j} → 대출 건수: {word}")

    except Exception as e:
        print(f"[오류 발생] ISBN: {j}, 오류 내용: {e}")
        df.loc[df['ISBN'] == j, 'LOAN_COUNT'] = None
        continue
```
