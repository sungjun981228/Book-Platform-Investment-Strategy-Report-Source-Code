# 월간 데이터 연간데이터로 통합

```python
import pandas as pd
import os

# ✅ 정확한 폴더 경로
folder_path = r"C:\Users\user\Desktop\연간 도서 데이터\yes24"

# ✅ .xlsx 확장자 파일만 추출 (대소문자 구분 없음)
excel_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.xlsx')]
print("📂 찾은 파일:", excel_files)

df_list = []

for file in excel_files:
    file_path = os.path.join(folder_path, file)
    print(f"📄 읽는 중: {file_path}")
    
    try:
        df = pd.read_excel(file_path)
        df['출처파일명'] = file  # 어디서 왔는지 추적 가능하게 추가
        df_list.append(df)
    except Exception as e:
        print(f"❌ {file} 읽기 실패: {e}")

if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)
    output_path = os.path.join(folder_path, 'yes24_통합_결과.xlsx')
    combined_df.to_excel(output_path, index=False)
    print(f"✅ 통합 완료! 저장 위치: {output_path}")
else:
    print("⚠️ 합칠 수 있는 파일이 없습니다.")
```

```python
import pandas as pd
import os

folder_path = r"C:\Users\user\Desktop\연간 도서 데이터\알라딘\월간 데이터"
file_list = [f for f in os.listdir(folder_path) if f.lower().endswith(('.xls', '.xlsx'))]

df_list = []

for file in file_list:
    file_path = os.path.join(folder_path, file)
    print(f"📄 처리 중: {file_path}")

    try:
        if file.lower().endswith('.xls'):
            try:
                # 1차 시도: 진짜 xls 형식이면
                df = pd.read_excel(file_path, engine="xlrd")
            except Exception:
                # 2차 시도: 사실상 CSV 파일일 경우
                df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
        else:
            df = pd.read_excel(file_path)

        df['출처파일명'] = file
        df_list.append(df)
    except Exception as e:
        print(f"❌ {file} 읽기 실패: {e}")

if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)
    output_path = os.path.join(folder_path, 'aladin_통합_결과.xlsx')
    combined_df.to_excel(output_path, index=False)
    print(f"✅ 저장 완료: {output_path}")
else:
    print("⚠️ 합칠 수 있는 파일이 없습니다.")
```

```python
import pandas as pd
import os

# ✅ 파일들이 저장된 폴더 경로 (여기에 CSV들이 들어 있음)
folder_path = r"C:\Users\user\Desktop\연간 도서 데이터\영풍"

# ✅ .csv 확장자만 필터링
csv_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.csv')]
print("📂 찾은 CSV 파일:", csv_files)

df_list = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    print(f"📄 처리 중: {file_path}")
    
    try:
        # 인코딩이 utf-8이 아니면 cp949(한글)로도 시도할 수 있음
        try:
            df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
        except UnicodeDecodeError:
            df = pd.read_csv(file_path, encoding='cp949', on_bad_lines='skip')

        df['출처파일명'] = file  # 어떤 파일에서 왔는지 기록
        df_list.append(df)
    except Exception as e:
        print(f"❌ {file} 읽기 실패: {e}")

# ✅ 병합 & 저장
if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)
    output_path = os.path.join(folder_path, 'ypbooks_통합_결과.xlsx')
    combined_df.to_excel(output_path, index=False)
    print(f"✅ 저장 완료: {output_path}")
else:
    print("⚠️ 합칠 수 있는 파일이 없습니다.")
```
