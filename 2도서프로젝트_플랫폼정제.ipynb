```python
import pandas as pd
import os

# âœ… íŒŒì¼ë“¤ì´ ì €ì¥ëœ í´ë” ê²½ë¡œ (ì—¬ê¸°ì— CSVë“¤ì´ ë“¤ì–´ ìˆìŒ)
folder_path = r"C:\Users\user\Desktop\ì—°ê°„ ë„ì„œ ë°ì´í„°\ì˜í’"

# âœ… .csv í™•ì¥ìë§Œ í•„í„°ë§
csv_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.csv')]
print("ğŸ“‚ ì°¾ì€ CSV íŒŒì¼:", csv_files)

df_list = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    print(f"ğŸ“„ ì²˜ë¦¬ ì¤‘: {file_path}")
    
    try:
        # ì¸ì½”ë”©ì´ utf-8ì´ ì•„ë‹ˆë©´ cp949(í•œê¸€)ë¡œë„ ì‹œë„í•  ìˆ˜ ìˆìŒ
        try:
            df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
        except UnicodeDecodeError:
            df = pd.read_csv(file_path, encoding='cp949', on_bad_lines='skip')

        df['ì¶œì²˜íŒŒì¼ëª…'] = file  # ì–´ë–¤ íŒŒì¼ì—ì„œ ì™”ëŠ”ì§€ ê¸°ë¡
        df_list.append(df)
    except Exception as e:
        print(f"âŒ {file} ì½ê¸° ì‹¤íŒ¨: {e}")

# âœ… ë³‘í•© & ì €ì¥
if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)
    output_path = os.path.join(folder_path, 'ypbooks_í†µí•©_ê²°ê³¼.xlsx')
    combined_df.to_excel(output_path, index=False)
    print(f"âœ… ì €ì¥ ì™„ë£Œ: {output_path}")
else:
    print("âš ï¸ í•©ì¹  ìˆ˜ ìˆëŠ” íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
```
