# 저자열 정규화

```python
import pandas as pd
import re

# 파일 불러오기
df = pd.read_csv('서울도서관_8차정제.csv')

# 정규화 함수 정의
def normalize_author(author):
    if pd.isna(author):
        return None
    author = str(author).strip()

    # 패턴: 이름 + (지음|지은이|발행|편|엮음|옮김) 중 하나
    match = re.match(r'(.*?)(지음|지은이|발행|편|엮음|옮김)', author)
    if match:
        return match.group(1).strip()

    # 그 외에는 그대로 (예: 이름만 있는 경우)
    return author

# AUTHOR 열 정규화
df['AUTHOR_normalized'] = df['AUTHOR'].apply(normalize_author)

# 결과 미리보기
print(df[['AUTHOR', 'AUTHOR_normalized']].head())

```
