```python
df = pd.read_csv('리뷰최종_교보베스트.csv')
```

```python
import pandas as pd

# 1. 리뷰 수 (review 개수)
review_counts = df.groupby('isbn')['review'].count().rename('리뷰수')

# 2. 감성 분석 결과 (긍정, 부정, 감성점수)
sentiment_counts = df.groupby(['isbn', 'sentiment']).size().unstack(fill_value=0)
sentiment_counts['감성점수'] = sentiment_counts.get('긍정', 0) - sentiment_counts.get('부정', 0)

# 3. 평점 평균
rating_means = df.groupby('isbn')['rating'].mean().rename('평균평점')

# 4. 세 결과 병합 (join으로 하나씩 붙이기)
result = sentiment_counts.join([review_counts, rating_means])

# 5. 열 순서 정리 (선택 사항)
result = result[['긍정', '부정', '감성점수', '리뷰수', '평균평점']]

# 결과 확인
result
```

```python
# 1. 감성 관련 열 이름 변경
result_renamed = result.rename(columns={
    '긍정': 'kyobo_po',
    '부정': 'kyobo_ne',
    '감성점수': 'kyobo_pnscore',
    '리뷰수': 'kyobo_review',
    '평균평점': 'kyobo_rating'
})

# 2. 'isbn' 인덱스를 'ISBN' 열로 변환
result_renamed = result_renamed.reset_index().rename(columns={'isbn': 'ISBN'})

# 3. df_best와 병합
df_merged = pd.merge(df_best, result_renamed, on='ISBN', how='left')

# 4. kyobo_rating 소수점 둘째 자리로 반올림
if 'kyobo_rating' in df_merged.columns:
    df_merged['kyobo_rating'] = df_merged['kyobo_rating'].round(2)

# 5. 감성 관련 열 중 NaN을 0으로 채우기
columns_to_fill = ['kyobo_po', 'kyobo_ne', 'kyobo_pnscore', 'kyobo_review', 'kyobo_rating']
for col in columns_to_fill:
    if col in df_merged.columns:
        df_merged[col] = df_merged[col].fillna(0)

# lANG 열 제거
df_merged.drop(['LANG', 'best_count'], axis=1, inplace=True)

# 1. 가중치 지정
weight_pnscore = 0.2
weight_review = 0.5
weight_rating = 0.3

# 2. 세 열이 모두 존재하는지 체크하고 계산
if all(col in df_merged.columns for col in ['kyobo_pnscore', 'kyobo_review', 'kyobo_rating']):
    df_merged['kyobo_score'] = (
        df_merged['kyobo_pnscore'] * weight_pnscore +
        df_merged['kyobo_review'] * weight_review +
        df_merged['kyobo_rating'] * weight_rating
    )
else:
    print("필요한 열이 하나 이상 존재하지 않습니다.")

if 'kyobo_score' in df_merged.columns:
    df_merged['kyobo_score'] = df_merged['kyobo_score'].round(2)


# 결과 확인
df_merged
```

```python
import pandas as pd

# ✅ 함수: 감성 요약 결과 만들기
def make_sentiment_result(df):
    review_counts = df.groupby('isbn')['review'].count().rename('리뷰수')
    sentiment_counts = df.groupby(['isbn', 'sentiment']).size().unstack(fill_value=0)
    sentiment_counts['감성점수'] = sentiment_counts.get('긍정', 0) - sentiment_counts.get('부정', 0)
    rating_means = df.groupby('isbn')['rating'].mean().rename('평균평점')
    result = sentiment_counts.join([review_counts, rating_means])
    result = result[['긍정', '부정', '감성점수', '리뷰수', '평균평점']]
    return result

# ✅ 함수: df_best와 감성 result 병합
def merge_sentiment_to_best(df_best, result, source_name):
    renamed = result.rename(columns={
        '긍정': f'{source_name}_po',
        '부정': f'{source_name}_ne',
        '감성점수': f'{source_name}_pnscore',
        '리뷰수': f'{source_name}_review',
        '평균평점': f'{source_name}_rating'
    }).reset_index().rename(columns={'isbn': 'ISBN'})
    
    merged = pd.merge(df_best, renamed, on='ISBN', how='left')

    if f'{source_name}_rating' in merged.columns:
        merged[f'{source_name}_rating'] = merged[f'{source_name}_rating'].round(2)

    cols_to_fill = [
        f'{source_name}_po',
        f'{source_name}_ne',
        f'{source_name}_pnscore',
        f'{source_name}_review',
        f'{source_name}_rating'
    ]
    for col in cols_to_fill:
        if col in merged.columns:
            merged[col] = merged[col].fillna(0)

    if all(col in merged.columns for col in [f'{source_name}_pnscore', f'{source_name}_review', f'{source_name}_rating']):
        merged[f'{source_name}_score'] = (
            merged[f'{source_name}_pnscore'] * 0.2 +
            merged[f'{source_name}_review'] * 0.5 +
            merged[f'{source_name}_rating'] * 0.3
        ).round(2)
    else:
        print(f"[{source_name}] 필요한 열이 하나 이상 존재하지 않습니다.")

    return merged
```

```python
# 0. 베스트 원본 불러오기 (df_best)
df_best = pd.read_csv('비베스트셀러_랜덤최종.csv')  # 너가 병합 기준으로 쓸 파일 이름
df_best.drop(['LANG', 'best_count'], axis=1, inplace=True)

# 1. 각 리뷰 데이터 불러오기
df_kyobo = pd.read_csv('리뷰최종_교보비베스트.csv')
df_aladin = pd.read_csv('리뷰최종_알라딘비베스트.csv')
df_yp = pd.read_csv('리뷰최종_영풍비베스트.csv')
df_yes24 = pd.read_csv('리뷰최종_예사비베스트.csv')

# 2. 각 감성 결과 만들기
kyobo_result = make_sentiment_result(df_kyobo)
aladin_result = make_sentiment_result(df_aladin)
yp_result = make_sentiment_result(df_yp)
yes24_result = make_sentiment_result(df_yes24)

# 3. 병합 순차 실행
df_best = merge_sentiment_to_best(df_best, kyobo_result, 'kyobo')
df_best = merge_sentiment_to_best(df_best, aladin_result, 'aladin')
df_best = merge_sentiment_to_best(df_best, yp_result, 'yp')
df_best = merge_sentiment_to_best(df_best, yes24_result, 'yes24')

df_best

# ✅ 결과 저장 (선택)
#df_best.to_csv('통합감성_베스트.csv', index=False)
```

```python
def make_sentiment_result(df):
    # 1. 리뷰 수
    review_counts = df.groupby('isbn')['review'].count().rename('리뷰수')

    # 2. 긍정/부정 개수
    sentiment_counts = df.groupby(['isbn', 'sentiment']).size().unstack(fill_value=0)

    # 3. 감성점수: 긍정 비율 = 긍정 / (긍정 + 부정)
    sentiment_counts['감성점수'] = (
        sentiment_counts.get('긍정', 0) / 
        (sentiment_counts.get('긍정', 0) + sentiment_counts.get('부정', 0))
    ).round(2)

    # 4. 평균 평점
    rating_means = df.groupby('isbn')['rating'].mean().rename('평균평점')

    # 5. 병합
    result = sentiment_counts.join([review_counts, rating_means])
    result = result[['긍정', '부정', '감성점수', '리뷰수', '평균평점']]

    return result

# ✅ 함수: df_best와 감성 result 병합
def merge_sentiment_to_best(df_best, result, source_name):
    renamed = result.rename(columns={
        '긍정': f'{source_name}_po',
        '부정': f'{source_name}_ne',
        '감성점수': f'{source_name}_pnscore',
        '리뷰수': f'{source_name}_review',
        '평균평점': f'{source_name}_rating'
    }).reset_index().rename(columns={'isbn': 'ISBN'})
    
    merged = pd.merge(df_best, renamed, on='ISBN', how='left')

    if f'{source_name}_rating' in merged.columns:
        merged[f'{source_name}_rating'] = merged[f'{source_name}_rating'].round(2)

    cols_to_fill = [
        f'{source_name}_po',
        f'{source_name}_ne',
        f'{source_name}_pnscore',
        f'{source_name}_review',
        f'{source_name}_rating'
    ]
    for col in cols_to_fill:
        if col in merged.columns:
            merged[col] = merged[col].fillna(0)

    if all(col in merged.columns for col in [f'{source_name}_pnscore', f'{source_name}_review', f'{source_name}_rating']):
        merged[f'{source_name}_score'] = (
            merged[f'{source_name}_pnscore'] * 0.1 +
            merged[f'{source_name}_review'] * 0.7 +
            merged[f'{source_name}_rating'] * 0.2
        ).round(2)
    else:
        print(f"[{source_name}] 필요한 열이 하나 이상 존재하지 않습니다.")

    return merged

# 0. 베스트 원본 불러오기 (df_best)
df_best = pd.read_csv('비베스트셀러_랜덤최종.csv')  # 너가 병합 기준으로 쓸 파일 이름
#df_best.drop(['LANG', 'best_count'], axis=1, inplace=True)

# 1. 각 리뷰 데이터 불러오기
df_kyobo = pd.read_csv('리뷰최종_교보비베스트.csv')
df_aladin = pd.read_csv('리뷰최종_알라딘비베스트.csv')
df_yp = pd.read_csv('리뷰최종_영풍비베스트.csv')
df_yes24 = pd.read_csv('리뷰최종_예사비베스트.csv')

# 2. 각 감성 결과 만들기
kyobo_result = make_sentiment_result(df_kyobo)
aladin_result = make_sentiment_result(df_aladin)
yp_result = make_sentiment_result(df_yp)
yes24_result = make_sentiment_result(df_yes24)

# 3. 병합 순차 실행
df_best = merge_sentiment_to_best(df_best, kyobo_result, 'kyobo')
df_best = merge_sentiment_to_best(df_best, aladin_result, 'aladin')
df_best = merge_sentiment_to_best(df_best, yp_result, 'yp')
df_best = merge_sentiment_to_best(df_best, yes24_result, 'yes24')
```

```python
# 1. 플랫폼 점수 열 목록
platform_scores = ['kyobo_score', 'aladin_score', 'yes24_score', 'yp_score']

# 2. CLASS_NO별 평균 점수 계산
genre_score_by_platform = df_best.groupby('CLASS_NO')[platform_scores].mean().round(2)

# 3. 각 CLASS_NO별로 플랫폼 점수 정렬 및 순위 저장
ranked_results = []

for class_no, row in genre_score_by_platform.iterrows():
    sorted_scores = row.sort_values(ascending=False)
    ranked_results.append({
        'CLASS_NO': class_no,
        '1위': sorted_scores.index[0].replace('_score', ''),
        '1위점수': sorted_scores.iloc[0],
        '2위': sorted_scores.index[1].replace('_score', ''),
        '2위점수': sorted_scores.iloc[1],
        '3위': sorted_scores.index[2].replace('_score', ''),
        '3위점수': sorted_scores.iloc[2],
        '4위': sorted_scores.index[3].replace('_score', ''),
        '4위점수': sorted_scores.iloc[3],
    })

# 4. 데이터프레임으로 변환 (전체 CLASS_NO 포함)
ranked_df = pd.DataFrame(ranked_results)

# 5. CLASS_NO 기준으로 정렬해서 보기 좋게 출력
ranked_df = ranked_df.sort_values('CLASS_NO')

# 6. 출력
print(ranked_df)
```
