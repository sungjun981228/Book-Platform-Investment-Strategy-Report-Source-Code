# 정제방법

- page열, 아이디열 지우기 ✅
- 별점 숫자 통일 (1,2,3,4,5) : 반올림으로 정제 (보류)
- 중복리뷰 지우기 ✅
- 리뷰같지않은거 지우기 - 종이책 어쩌구저쩌구 써있는거, 숫자 등등 ✅
- 리뷰 누락 지우기 ✅
- 이모지, 특수문자, 줄바꿈 제거
- 나중에 다른 애들하고 컬럼명 통일시키기

### 데이터프레임 목록
- kb_review_best : 베스트셀러 리뷰
- kb_review_notbest : 비베스트셀러 리뷰

```python
import pandas as pd
```

# 베스트셀러

```python
kb_review_best = pd.read_csv('[교보] 베스트셀러 리뷰.csv')
kb_review_best.info()
```

## page 열 지우기

```python
kb_review_best = kb_review_best.drop(['page'], axis = 1)
```

## 중복 행 지우기

```python
kb_review_best = kb_review_best.drop_duplicates()
```

```python
kb_review_best
```

## 리뷰같지않은거 지우기

```python
import re

pattern = r"종이책.*\|\s*\d{4}\.\d{2}\.\d{2}\s*\|\s*신고/차단"
```

```python
kb_review_best = kb_review_best[~kb_review_best['review'].str.contains(pattern, na=False)]
```

```python
kb_review_best
```

## 누락리뷰 지우기

```python
kb_review_best = kb_review_best[~kb_review_best['review'].isna()]
```

## 유저아이디 지우기

```python
kb_review_best = kb_review_best.drop(['user_id'], axis = 1)
```

```python
kb_review_best
```

## 숫자로만 이루어진 리뷰 삭제

```python
kb_review_best = kb_review_best[~kb_review_best['review'].str.contains(r'^\d+$')]
```

```python
kb_review_best
```

## 이모지, 특수문자, 줄바꿈 제거

```python
import re

def cleansing(text):
    if pd.isnull(text):
        return text

    # 1. 줄바꿈 문자 제거
    text = text.replace('\n', ' ').replace('\r', ' ')

    # 2. 이모지 제거 (유니코드 기반)
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text)

    # 3. 특수문자 제거 (한글, 영문, 숫자, 공백만 남김)
    text = re.sub(r'[^가-힣a-zA-Z0-9\s]', '', text)

    # 4. 공백 여러 개 → 하나로 압축
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
kb_review_best['review'] = kb_review_best['review'].apply(cleansing)
```

```python
kb_review_best
```

## 별점 반올림

```python
kb_review_best['rating'] = kb_review_best['rating'].round().astype(int)
```

```python
kb_review_best
```

# 비베스트셀러

```python
kb_review_notbest = pd.read_csv('[교보] 비베스트셀러 리뷰.csv')
kb_review_notbest.info()
```

```python
# page열,유저아이디열 지우기

kb_review_notbest = kb_review_notbest.drop(['page'], axis = 1)
kb_review_notbest = kb_review_notbest.drop(['user_id'], axis = 1)
```

```python
# 중복 행 지우기

kb_review_notbest = kb_review_notbest.drop_duplicates()
```

```python
kb_review_notbest
```

```python
# 리뷰같지않은거 지우기

kb_review_notbest = kb_review_notbest[~kb_review_notbest['review'].str.contains("이벤트 신청: http://|원문보기: ", na=False)]
kb_review_notbest = kb_review_notbest[~kb_review_notbest['review'].str.contains(pattern, na=False)]
```

```python
# 누락리뷰 지우기

kb_review_notbest = kb_review_notbest[~kb_review_notbest['review'].isna()]
```

```python
kb_review_notbest
```

```python
# 숫자로만 이루어진 리뷰 삭제

kb_review_notbest = kb_review_notbest[~kb_review_notbest['review'].str.contains(r'^\d+$')]
```

```python
# 이모지, 특수문자, 줄바꿈 제거
kb_review_notbest['review'] = kb_review_notbest['review'].apply(cleansing)
kb_review_notbest
```

```python
# 별점 반올림
kb_review_notbest['rating'] = kb_review_notbest['rating'].round().astype(int)
```

```python
kb_review_notbest
```

```python
kb_review_best.to_csv("[교보] 베스트셀러 리뷰_1차정제.csv", index=False, encoding='utf-8-sig')
kb_review_notbest.to_csv("[교보] 비베스트셀러 리뷰_1차정제.csv", index=False, encoding='utf-8-sig')
```
