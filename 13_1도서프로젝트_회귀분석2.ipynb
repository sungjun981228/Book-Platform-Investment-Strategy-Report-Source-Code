# 정규분포 확인

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

score_columns = ['kyobo_score', 'aladin_score', 'yp_score', 'yes24_score']

for col in score_columns:
    plt.figure(figsize=(12, 5))
    
    # 히스토그램
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'{col} Histogram')

    # Q-Q plot
    plt.subplot(1, 2, 2)
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.title(f'{col} Q-Q Plot')
    
    plt.tight_layout()
    plt.show()
```

```python
from scipy.stats import shapiro

for col in score_columns:
    stat, p = shapiro(df[col])
    print(f'{col} - Shapiro-Wilk p-value: {p:.4f}')
```

# 점수 상위 20% 데이터 분리, 나중에 테스트용 데이터 분리

```python
# 1. best_score 열 생성 (4개 점수 합계)
df_fn['best_score'] = df_fn[['kyobo_score', 'aladin_score', 'yp_score', 'yes24_score']].sum(axis=1)

# 2. best_score 기준 상위 20% 커트라인 계산
threshold = df_fn['best_score'].quantile(0.80)

# 3. 상위 20% 데이터 추출
top_20_percent_df = df_fn[df_fn['best_score'] >= threshold]

top_20_percent_df
```

# 스케일링 진행
정규분포가 아니기 때문에 MinMax 스케일링 사용

```python
from sklearn.preprocessing import MinMaxScaler

# 스케일링할 열 목록
score_columns = ['kyobo_score', 'aladin_score', 'yp_score', 'yes24_score']

# MinMaxScaler 객체 생성
scaler = MinMaxScaler()
scaler.fit(df_fn[score_columns])
df_fn[score_columns] = scaler.transform(df_fn[score_columns])

df_fn
```
