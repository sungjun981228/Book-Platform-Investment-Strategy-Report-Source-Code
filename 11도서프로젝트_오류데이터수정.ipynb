```python
import pandas as pd

# 1) df의 CLASS_NO별 목표 개수 구하기
target_counts = df['CLASS_NO'].value_counts()

# 2) df1에서 목표 개수만큼씩 뽑아내기
#    - df1에 해당 CLASS_NO가 더 적으면 전부 유지하고 경고만 출력
sampled_chunks = []

for cls, n in target_counts.items():
    df1_cls = df1[df1['CLASS_NO'] == cls]

    if len(df1_cls) >= n:
        # 목표 개수만큼 무작위(랜덤 시드 고정) 추출
        sampled = df1_cls.sample(n=n, random_state=42)
    else:
        # 부족한 경우 경고 메시지
        print(f"[Warning] CLASS_NO {cls}: df1에 {len(df1_cls)}개밖에 없어 {n}개로 맞출 수 없습니다.")
        sampled = df1_cls          # 있는 만큼 유지

    sampled_chunks.append(sampled)

# 3) 조각들을 합쳐서 최종 df1'
df1_matched = pd.concat(sampled_chunks).reset_index(drop=True)

# 4) 검증: 행 수와 CLASS_NO 분포 확인
print("df  행 수 :", len(df))
print("df1 행 수 :", len(df1_matched))
print("\nCLASS_NO별 개수 비교")
print(pd.concat(
        [df['CLASS_NO'].value_counts().sort_index().rename("df"),
         df1_matched['CLASS_NO'].value_counts().sort_index().rename("df1_matched")],
        axis=1
     ))
```

# 감성점수 다시 산출
- 가중치 잘못 부여되서 다시 산출

```python
def make_sentiment_result(df):
    # 1. 리뷰 수
    review_counts = df.groupby('isbn')['review'].count().rename('리뷰수')

    # 2. 긍정/부정 개수
    sentiment_counts = df.groupby(['isbn', 'sentiment']).size().unstack(fill_value=0)

    # 3. 감성점수: 긍정 비율 = 긍정 / (긍정 + 부정)
    sentiment_counts['감성점수'] = (
        sentiment_counts.get('긍정', 0) / 
        (sentiment_counts.get('긍정', 0) + sentiment_counts.get('부정', 0))
    ).round(2)

    # 4. 평균 평점
    rating_means = df.groupby('isbn')['rating'].mean().rename('평균평점')

    # 5. 병합
    result = sentiment_counts.join([review_counts, rating_means])
    result = result[['긍정', '부정', '감성점수', '리뷰수', '평균평점']]

    return result

# ✅ 함수: df_best와 감성 result 병합
def merge_sentiment_to_best(df_best, result, source_name):
    renamed = result.rename(columns={
        '긍정': f'{source_name}_po',
        '부정': f'{source_name}_ne',
        '감성점수': f'{source_name}_pnscore',
        '리뷰수': f'{source_name}_review',
        '평균평점': f'{source_name}_rating'
    }).reset_index().rename(columns={'isbn': 'ISBN'})
    
    merged = pd.merge(df_best, renamed, on='ISBN', how='left')

    if f'{source_name}_rating' in merged.columns:
        merged[f'{source_name}_rating'] = merged[f'{source_name}_rating'].round(2)

    cols_to_fill = [
        f'{source_name}_po',
        f'{source_name}_ne',
        f'{source_name}_pnscore',
        f'{source_name}_review',
        f'{source_name}_rating'
    ]
    for col in cols_to_fill:
        if col in merged.columns:
            merged[col] = merged[col].fillna(0)

    if all(col in merged.columns for col in [f'{source_name}_pnscore', f'{source_name}_review', f'{source_name}_rating']):
        merged[f'{source_name}_score'] = (
            merged[f'{source_name}_pnscore'] * 0.1 +
            merged[f'{source_name}_review'] * 0.7 +
            merged[f'{source_name}_rating'] * 0.2
        ).round(2)
    else:
        print(f"[{source_name}] 필요한 열이 하나 이상 존재하지 않습니다.")

    return merged

# 0. 베스트 원본 불러오기 (df_best)
df_best = pd.read_csv('베스트셀러_최종버전6.csv')  # 너가 병합 기준으로 쓸 파일 이름
#df_best.drop(['LANG', 'best_count'], axis=1, inplace=True)

# 1. 각 리뷰 데이터 불러오기
df_kyobo = pd.read_csv('리뷰최종_교보베스트.csv')
df_aladin = pd.read_csv('리뷰최종_알라딘베스트.csv')
df_yp = pd.read_csv('리뷰최종_영풍베스트.csv')
df_yes24 = pd.read_csv('리뷰최종_예사베스트.csv')

# 2. 각 감성 결과 만들기
kyobo_result = make_sentiment_result(df_kyobo)
aladin_result = make_sentiment_result(df_aladin)
yp_result = make_sentiment_result(df_yp)
yes24_result = make_sentiment_result(df_yes24)

# 3. 병합 순차 실행
df_best = merge_sentiment_to_best(df_best, kyobo_result, 'kyobo')
df_best = merge_sentiment_to_best(df_best, aladin_result, 'aladin')
df_best = merge_sentiment_to_best(df_best, yp_result, 'yp')
df_best = merge_sentiment_to_best(df_best, yes24_result, 'yes24')
```

```python
import pandas as pd
def make_sentiment_result(df):
    # 1. 리뷰 수
    review_counts = df.groupby('isbn')['review'].count().rename('리뷰수')

    # 2. 긍정/부정 개수
    sentiment_counts = df.groupby(['isbn', 'sentiment']).size().unstack(fill_value=0)

    # 3. 감성점수: 긍정 비율 = 긍정 / (긍정 + 부정)
    sentiment_counts['감성점수'] = (
        sentiment_counts.get('긍정', 0) / 
        (sentiment_counts.get('긍정', 0) + sentiment_counts.get('부정', 0))
    ).round(2)

    # 4. 평균 평점
    rating_means = df.groupby('isbn')['rating'].mean().rename('평균평점')

    # 5. 병합
    result = sentiment_counts.join([review_counts, rating_means])
    result = result[['긍정', '부정', '감성점수', '리뷰수', '평균평점']]

    return result

# ✅ 함수: df_best와 감성 result 병합
def merge_sentiment_to_best(df_best, result, source_name):
    renamed = result.rename(columns={
        '긍정': f'{source_name}_po',
        '부정': f'{source_name}_ne',
        '감성점수': f'{source_name}_pnscore',
        '리뷰수': f'{source_name}_review',
        '평균평점': f'{source_name}_rating'
    }).reset_index().rename(columns={'isbn': 'ISBN'})
    
    merged = pd.merge(df_best, renamed, on='ISBN', how='left')

    if f'{source_name}_rating' in merged.columns:
        merged[f'{source_name}_rating'] = merged[f'{source_name}_rating'].round(2)

    cols_to_fill = [
        f'{source_name}_po',
        f'{source_name}_ne',
        f'{source_name}_pnscore',
        f'{source_name}_review',
        f'{source_name}_rating'
    ]
    for col in cols_to_fill:
        if col in merged.columns:
            merged[col] = merged[col].fillna(0)

    if all(col in merged.columns for col in [f'{source_name}_pnscore', f'{source_name}_review', f'{source_name}_rating']):
        merged[f'{source_name}_score'] = (
            merged[f'{source_name}_pnscore'] * 0.5 +
            merged[f'{source_name}_review'] * 0.05 +
            merged[f'{source_name}_rating'] * 0.45
        ).round(2)
    else:
        print(f"[{source_name}] 필요한 열이 하나 이상 존재하지 않습니다.")

    return merged

# 0. 베스트 원본 불러오기 (df_best)
df_best = pd.read_csv('비베스트셀러_최종버전9.csv')  # 너가 병합 기준으로 쓸 파일 이름
df_best.drop(['kyobo_po', 'kyobo_ne', 'kyobo_pnscore', 'kyobo_review',
       'kyobo_rating', 'kyobo_score', 'aladin_po', 'aladin_ne',
       'aladin_pnscore', 'aladin_review', 'aladin_rating', 'aladin_score',
       'yp_po', 'yp_ne', 'yp_pnscore', 'yp_review', 'yp_rating', 'yp_score',
       'yes24_po', 'yes24_ne', 'yes24_pnscore', 'yes24_review', 'yes24_rating',
       'yes24_score'], axis=1, inplace=True)
#df_best.drop(['LANG', 'best_count'], axis=1, inplace=True)

# 1. 각 리뷰 데이터 불러오기
df_kyobo = pd.read_csv('리뷰최종_교보비베스트.csv')
df_aladin = pd.read_csv('리뷰최종_알라딘비베스트.csv')
df_yp = pd.read_csv('리뷰최종_영풍비베스트.csv')
df_yes24 = pd.read_csv('리뷰최종_예사비베스트.csv')

# 2. 각 감성 결과 만들기
kyobo_result = make_sentiment_result(df_kyobo)
aladin_result = make_sentiment_result(df_aladin)
yp_result = make_sentiment_result(df_yp)
yes24_result = make_sentiment_result(df_yes24)

# 3. 병합 순차 실행
df_best = merge_sentiment_to_best(df_best, kyobo_result, 'kyobo')
df_best = merge_sentiment_to_best(df_best, aladin_result, 'aladin')
df_best = merge_sentiment_to_best(df_best, yp_result, 'yp')
df_best = merge_sentiment_to_best(df_best, yes24_result, 'yes24')
```

```python
# 평가점수 1점당 각 요소 얼마나 차이나는지
# 1. score 10점대와 11점대 비교 그룹 만들기
group_10 = df_best[(df_best['kyobo_score'] >= 9.9) & (df_best['kyobo_score'] <= 10.1)]
group_11 = df_best[(df_best['kyobo_score'] >= 10.9) & (df_best['kyobo_score'] <= 11.1)]

# 2. 평균값 비교
score_compare = pd.DataFrame({
    'pnscore': [group_10['kyobo_pnscore'].mean(), group_11['kyobo_pnscore'].mean()],
    'review': [group_10['kyobo_review'].mean(), group_11['kyobo_review'].mean()],
    'rating': [group_10['kyobo_rating'].mean(), group_11['kyobo_rating'].mean()]
}, index=['score_10', 'score_11'])

# 3. 요소별 차이 계산
score_compare.loc['차이'] = score_compare.loc['score_11'] - score_compare.loc['score_10']

# 4. 각 요소별 기여도 계산 (가중치 반영)
contrib = {
    'pnscore': score_compare.loc['차이', 'pnscore'] * 0.5,
    'review': score_compare.loc['차이', 'review'] * 0.05,
    'rating': score_compare.loc['차이', 'rating'] * 0.45
}
contrib['합계'] = sum(contrib.values())

# 5. 출력
print("\n📊 요소별 score 변화 기여도")
for k, v in contrib.items():
    print(f"{k}: {v:.3f}")
```
