```python
import pandas as pd
import re
from datetime import datetime

# 현재 연도
CURRENT_YEAR = datetime.now().year

def extract_dates(period_str):
    if not isinstance(period_str, str):
        return pd.Series([None, None])

    # 패턴 1: YYYY/MM/DD ~ YYYY/MM/DD 또는 YYYY.MM.DD ~ YYYY.MM.DD
    match_full = re.findall(r'(\d{4}[./]\d{2}[./]\d{2})', period_str)
    if len(match_full) == 2:
        start = match_full[0].replace('.', '/')
        end = match_full[1].replace('.', '/')
        return pd.Series([start, end])

    # 패턴 2: YYYY/MM/DD ~ (종료일 없음)
    match_start_only = re.match(r'(\d{4})[./](\d{2})[./](\d{2})\s*~', period_str)
    if match_start_only:
        y, m, d = match_start_only.groups()
        start = f"{y}/{m}/{d}"

        # "한정수량", "재고소진", "재고 소진" 키워드 있으면
        if "한정수량" in period_str or "재고소진" in period_str or "재고 소진" in period_str:
            end = "재고소진시 종료"
        else:
            end = "진행 중"

        return pd.Series([start, end])

    # 패턴 3: 5월 21일 ~ 형식
    match_korean = re.match(r'(\d{1,2})월\s*(\d{1,2})일\s*~', period_str)
    if match_korean:
        m, d = map(int, match_korean.groups())
        start = f"{CURRENT_YEAR}/{m:02d}/{d:02d}"
        end = "진행 중"
        return pd.Series([start, end])

    return pd.Series([None, None])

# 적용
df_all[['start_date', 'end_date']] = df_all['기간'].apply(extract_dates)
```

# 플랫폼별 이벤트
- 각 플랫폼 데이터 열 생성
- 알라딘 데이터 증정 내용 전부 누락값, 해결책 필요함 -> 정보 없음으로 채움, 알라딘 데이터 수가 너무 많음 3/2

- 일단 나중에 해야할듯
